```python
import numpy as np
def matmul(A, B):
    """
    Multiply two matrices A (m x p) and B (p x n) using triple loop.
    Returns C = A @ B (m x n)
    """
    # Get shapes
    m, p1 = A.shape
    p2, n = B.shape
    
    # Check compatibility
    if p1 != p2:
        raise ValueError(f"Inner dimensions don't match: {p1} != {p2}")
    
    # Initialize result matrix with zeros
    C = np.zeros((m, n))
    
    # Triple loop: for each row of A, each column of B, dot product over p
    for i in range(m):          # rows of A / C
        for j in range(n):      # columns of B / C
            for k in range(p1): # columns of A / rows of B
                C[i, j] += A[i, k] * B[k, j]
    
    return C
# ───────────────────────────────────────────────────────────────
# Test cases
# ───────────────────────────────────────────────────────────────
# Test 1: 2×2 × 2×2
A1 = np.array([[1, 2],
               [3, 4]])
B1 = np.array([[5, 6],
               [7, 8]])
print("A1 @ B1 (manual):")
print(matmul(A1, B1))
print("NumPy @:")
print(A1 @ B1)
# Test 2: 3×3 × 3×2
A2 = np.arange(1, 10).reshape(3, 3)
B2 = np.array([[1, 2], [3, 4], [5, 6]])
print("\nA2 @ B2 (manual):")
print(matmul(A2, B2))
# Test 3: 2×3 × 3×1 (matrix-vector)
A3 = np.array([[1, 0, 1],
               [2, 1, 0]])
B3 = np.array([[4], [5], [6]])
print("\nA3 @ B3 (manual):")
print(matmul(A3, B3))

```

```python
import numpy as np

def matmul(A,B):
    m,p1 = A.shape
    p2,n = B.shape
    
    if p1 != p2:
        raise ValueError ('These matrices cannot be multiplied, p1 is not equal to p2')
        
    C = np.zeros((m,n))
    
    for i in range(m):
        for j in range(n):
            for k in range(p1):
                C[i,j] += A[i,k] * B[k,j] 

    return C
A = np.array([[1, 2, 3],
              [4, 5, 6]])           # 2×3

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])            # 3×2
print("A @ B (manual):")
print(matmul(A, B))
print("NumPy @:")
print(A @ B)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_transformation(A, title="Transformation"):
    """
    Visualize how matrix A transforms the unit circle and a grid.
    Overlays eigenvectors if they exist and are real.
    """
    # ─── Generate points ────────────────────────────────────────
    # Unit circle
    theta = np.linspace(0, 2*np.pi, 100)
    circle = np.vstack([np.cos(theta), np.sin(theta)])  # 2×100

    # Grid of points (square [-1.5,1.5]×[-1.5,1.5])
    x = np.linspace(-1.5, 1.5, 15)
    y = np.linspace(-1.5, 1.5, 15)
    X, Y = np.meshgrid(x, y)
    grid = np.vstack([X.ravel(), Y.ravel()])  # 2×N

    # ─── Apply transformation ────────────────────────────────────
    circle_transformed = A @ circle
    grid_transformed   = A @ grid

    # ─── Compute eigenvectors (if possible) ──────────────────────
    try:
        eigenvalues, eigenvectors = np.linalg.eig(A)
        # Only keep real eigenvectors
        real_mask = np.isreal(eigenvalues)
        real_eigvals = eigenvalues[real_mask].real
        real_eigvecs = eigenvectors[:, real_mask].real

        # Normalize vectors to length ~1 for plotting
        norms = np.linalg.norm(real_eigvecs, axis=0)
        real_eigvecs = real_eigvecs / (norms + 1e-10)  # avoid div by zero
    except:
        real_eigvecs = np.empty((2, 0))  # no real eigenvectors

    # ─── Plotting ────────────────────────────────────────────────
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)

    # Original
    ax1.plot(circle[0], circle[1], 'b-', lw=1.5, label='Unit circle')
    ax1.scatter(grid[0], grid[1], s=10, color='lightblue', alpha=0.6)
    ax1.set_title("Before (original)")
    ax1.set_aspect('equal')
    ax1.grid(True, alpha=0.3)
    ax1.axhline(0, color='gray', lw=0.5)
    ax1.axvline(0, color='gray', lw=0.5)

    # Transformed
    ax2.plot(circle_transformed[0], circle_transformed[1], 'r-', lw=1.5, label='Transformed')
    ax2.scatter(grid_transformed[0], grid_transformed[1], s=10, color='salmon', alpha=0.6)
    ax2.set_title(f"After {title}")
    ax2.set_aspect('equal')
    ax2.grid(True, alpha=0.3)
    ax2.axhline(0, color='gray', lw=0.5)
    ax2.axvline(0, color='gray', lw=0.5)

    # Overlay eigenvectors (arrows from origin)
    if real_eigvecs.shape[1] > 0:
        for idx in range(real_eigvecs.shape[1]):
            vec = real_eigvecs[:, idx]
            scale = 1.2  # make arrows longer for visibility
            ax1.quiver(0, 0, vec[0]*scale, vec[1]*scale,
                       color='green', scale_units='xy', scale=1, width=0.008,
                       label=f"eigvec {idx+1}" if idx == 0 else None)
            ax2.quiver(0, 0, vec[0]*scale, vec[1]*scale,
                       color='green', scale_units='xy', scale=1, width=0.008)

    ax1.legend()
    ax2.legend()
    plt.tight_layout()
    plt.show()

# ───────────────────────────────────────────────────────────────
# Try different matrices!
# ───────────────────────────────────────────────────────────────


# Example 4: Reflection over y=x
reflection = np.array([[1, 1],
                       [1, -1]])
plot_transformation(reflection, "Reflection y=x")

# # Example 5: Projection onto x-axis (rank 1)
# projection = np.array([[1, 0],
#                        [0, 0]])
# plot_transformation(projection, "Projection to x-axis")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ─── Pick a symmetric matrix from your visualizer ─────────────────
# Example: the one with perpendicular stretches
A = np.array([[2, 3], 
              [-1, 4]])

# ─── Compute SVD ──────────────────────────────────────────────────
U, S, Vt = np.linalg.svd(A, full_matrices=True)

print("U (left singular vectors):\n", np.round(U, 3))
print("\nSingular values (Σ diagonal):\n", np.round(S, 3))
print("\nVt (V transpose, right singular vectors):\n", np.round(Vt, 3))

# Reconstruct full A
Sigma = np.zeros((U.shape[1], Vt.shape[0]))
Sigma[:len(S), :len(S)] = np.diag(S)
reconstructed = U @ Sigma @ Vt
print("\nReconstructed A (should match original):\n", np.round(reconstructed, 3))

# ─── Low-rank approx example (keep top 1 singular value) ─────────
k = 1
U_k = U[:, :k]
Sigma_k = np.diag(S[:k])
Vt_k = Vt[:k, :]
low_rank = U_k @ Sigma_k @ Vt_k
print("\nLow-rank approx (k=1):\n", np.round(low_rank, 3))

# ─── Visualize the full SVD pipeline on unit circle ───────────────
theta = np.linspace(0, 2*np.pi, 100)
circle = np.vstack([np.cos(theta), np.sin(theta)])  # 2×100

# Step 1: Apply Vᵀ (rotate input space)
after_Vt = Vt @ circle

# Step 2: Apply Σ (perpendicular stretch)
Sigma_full = np.zeros((2, 2))
Sigma_full[:len(S), :len(S)] = np.diag(S)
after_Sigma = Sigma_full @ after_Vt

# Step 3: Apply U (rotate output space)
after_U = U @ after_Sigma

# ─── Plot all stages ──────────────────────────────────────────────
fig, axs = plt.subplots(1, 4, figsize=(16, 4), sharex=True, sharey=True)

stages = [
    (circle, 'Original circle', 'blue'),
    (after_Vt, 'After Vᵀ (input rotation)', 'orange'),
    (after_Sigma, 'After Σ (stretch)', 'green'),
    (after_U, 'After U (output rotation)', 'red')
]

for ax, (pts, lbl, color) in zip(axs, stages):
    ax.plot(pts[0], pts[1], color=color, lw=1.5)
    ax.set_title(lbl)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(0, color='gray', lw=0.5)
    ax.axvline(0, color='gray', lw=0.5)

plt.tight_layout()
plt.show()
```
